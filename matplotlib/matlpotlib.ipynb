```python
import matplotlib.pyplot as matplot
import numpy as np
```

```python
%matplotlib inline 
##jupiter notebook özel
```

```python
yas_list=[6,8,13,19,2,16,19,64,52,13,16,12]
kilo_list=[20,60,80,10,65,45,90,60,70,75,50,40]
np_Yas_list=np.array(yas_list)
np_kilo_list=np.array(kilo_list)
```

```python
matplot.plot(np_Yas_list,np_kilo_list,"g",)#y-sarı r-kırmızı fln
```

```python
matplot.xlabel("yas verileri")
matplot.ylabel("kilo verileri")
matplot.plot(np_Yas_list,np_kilo_list,"g",)#y-sarı r-kırmızı fln
# eğer başka bir editörde yazıyorsarsak matplot.show()
```

```python
numpy_dizisi=np.linspace(0,20,10)
numpy_dizisi_index=np.array(range(len(numpy_dizisi)))
matplot.plot(numpy_dizisi,numpy_dizisi_index,"r")
```

```python
matplot.plot(numpy_dizisi,numpy_dizisi_index,"r--")
```

```python
matplot.plot(numpy_dizisi,numpy_dizisi_index,"r*-")
```

```python
matplot.subplot(1,2,1)#kaç sıra ,kaç grafik, bu satırda hangisi çizilcek
matplot.plot(numpy_dizisi**2,numpy_dizisi_index,"r*-")
matplot.subplot(1,2,2)
matplot.plot(numpy_dizisi_index,numpy_dizisi**2,"g-")
```

## figür
matplotlib kütüphanesinden => matplot.figure()
.figure([sol_başlangıçN,alt_başlangıç,genişlik,uzunluk])

```python
benimfigür=matplot.figure()
```

```python
benimfigür=matplot.figure()
figüraxes=benimfigür.add_axes([0.1,0.1,0.3,0.3])
figüraxes.plot(numpy_dizisi_index,numpy_dizisi**2,"g")
figüraxes.set_xlabel("x ekseni")
figüraxes.set_ylabel("y ekseni")
figüraxes.set_title("başlık")
```

```python
figure2=matplot.figure()
eksen1= figure2.add_axes([0.05,0.1,0.7,0.7])
eksen2=figure2.add_axes([0.1,0.45,0.3,0.3])
eksen1.plot(numpy_dizisi,numpy_dizisi_index,"g")
eksen1.set_xlabel("x ekseni")
eksen1.set_ylabel("y ekseni")
eksen1.set_title("baslık")
eksen2.plot(numpy_dizisi_index,numpy_dizisi,"b")
eksen2.set_xlabel("x ekseni")
eksen2.set_ylabel("y ekseni")
eksen2.set_title("başlık")
```

```python
#subplot'lar ile de bu işelemler yapılabilir
(benimfigür,benimaxem)=matplot.subplots()#tuple 
benimaxem.plot(numpy_dizisi,numpy_dizisi_index)
```
