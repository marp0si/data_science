```python
import numpy as np
import matplotlib.pyplot as matplot
%matplotlib inline
```

```python
numpydizisi1=np.linspace(0,10,20)
numpydizisi2=numpydizisi1**2
```

```python
(figür1,eksen1)=matplot.subplots()
eksen1.plot(numpydizisi1,numpydizisi2,"g")#renklerin hex kodları kullanarak çok fazla renk var
eksen1.plot(numpydizisi1**2,numpydizisi2,color="#DC143C",alpha=0.5)#alpha saylamlık 1 en fazla
```

```python
(figür2,eksen2)=matplot.subplots()
eksen2.plot(numpydizisi1,numpydizisi2-5,"red")
eksen2.plot(numpydizisi1,numpydizisi2,"orange",linewidth=6.2)#kalınlık default=2
eksen2.plot(numpydizisi1,numpydizisi2+4,color="#A52A2A",linestyle="--")#linestyle --,:,-*
eksen2.plot(numpydizisi1,numpydizisi2+10,marker="*",color="#000000",markerfacecolor="blue",markersize=4)
```

## scatter


```python
matplot.scatter(numpydizisi1,numpydizisi2)
```

## histogram

```python
numpydizisi3=np.random.randint(0,100,60)
matplot.hist(numpydizisi3)
```

```python
matplot.boxplot(numpydizisi3)
```
